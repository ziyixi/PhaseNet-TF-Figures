To plot the gamma 1D example, we need to know in which 10min window, there are most frequency picks. We can use the following code to find out.

```python
import pandas as pd
from tqdm import tqdm
```

```python
assignments=pd.read_csv("/Users/ziyixi/Library/CloudStorage/OneDrive-MichiganStateUniversity/Paper/PhaseNetTF_myturn/PhaseNet-TF-Figures/phasenettf/data/catalog/continious_associated_assignment.csv",skiprows=1,names=
                        ["id",	"date" ,"time"	,"amp"	,"type"	,"prob"	,"event_index"	,"gamma_score"],sep="\s+")
assignments["time"]=assignments[['date', 'time']].apply(lambda x: ' '.join(x), axis=1)
assignments["time"]=pd.to_datetime(assignments["time"])
```

```python
assignments
```

```python
# check in which 10 minutes the most picks are made, the 10 minutes window start at integer minutes such as 0min, 10min, 20min, 30min, 40min, 50min
# below is the code
start_time=pd.to_datetime("2009-11-01 00:00:00")
end_time=pd.to_datetime("2011-01-01 00:00:00")
time_range=pd.date_range(start=start_time,end=end_time,freq="10min")

frequency=[0]*(len(time_range)-1)
def check_frequency(row):
    # get row.time position in the frequency list
    time=row.time
    diff_to_start=time-time_range[0]
    index=int(diff_to_start.total_seconds()/600)
    frequency[index]+=1
assignments.apply(check_frequency,axis=1);
```

```python
# get the top 10 time range and corresponding frequency
top_10_frequency=sorted(frequency,reverse=True)[:10]
top_10_time_range=[]
for i in range(10):
    index=frequency.index(top_10_frequency[i])
    top_10_time_range.append(time_range[index:index+2])

for i in range(10):
    print("top {} time range is {} to {}, frequency is {}".format(i+1,top_10_time_range[i][0],top_10_time_range[i][1],top_10_frequency[i]))
```
