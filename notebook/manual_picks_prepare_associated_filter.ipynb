For the manual picks test, this notebook is designed to filter the catalog by 10 events.

```python
import pandas as pd
```

```python
assignment=pd.read_csv("/Users/ziyixi/Library/CloudStorage/OneDrive-MichiganStateUniversity/Paper/PhaseNetTF_myturn/PhaseNet-TF-Figures/phasenettf/data/catalog/manual_picks/manual_phase_picks.assignments.csv",sep=r"\s+")
```

```python
assignment.head()
```

```python
catalog=pd.read_csv("/Users/ziyixi/Library/CloudStorage/OneDrive-MichiganStateUniversity/Paper/PhaseNetTF_myturn/PhaseNet-TF-Figures/phasenettf/data/catalog/manual_picks/manual_phase_picks.catalog.csv",sep=r"\s+")
```

```python
catalog.head()
```

```python
# filter event_index with more than 10 assignment rows
event_index=assignment.groupby("event_index").count()
event_index=event_index[event_index["event_id"]>10]
event_index=set(event_index.index.tolist())

catalog=catalog[catalog["event_index"].isin(event_index)]
```

```python
catalog.head()
catalog=catalog[["event_index","latitude","longitude","z(km)"]]
catalog.to_csv("/Users/ziyixi/Library/CloudStorage/OneDrive-MichiganStateUniversity/Paper/PhaseNetTF_myturn/PhaseNet-TF-Figures/phasenettf/data/catalog/manual_picks/manual_phase_picks.association_filtered.csv",index=False)
```
