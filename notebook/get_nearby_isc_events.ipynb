This notebook is designed to get nearby events, so I can plot them all together.

```python
import pandas as pd
```

```python
data=pd.read_csv('/Users/ziyixi/Library/CloudStorage/OneDrive-MichiganStateUniversity/Paper/PhaseNetTF_myturn/PhaseNet-TF-Figures/phasenettf/data/catalog/tonga_catalog_updated_2023_0426.csv')
data["time"]=pd.to_datetime(data["time"])
data=data[data["ISCid"]>0]
```

```python
data
```

```python
# find two nearby events in time with origin time difference less than 10 seconds
pairs=[]
for i in range(len(data)):
    for j in range(i+1,len(data)):
        if abs((data.iloc[i]["time"]-data.iloc[j]["time"]).total_seconds())<200:
            pairs.append([data.iloc[i]["originid"],data.iloc[j]["originid"],data.iloc[i]["time"],data.iloc[j]["time"]])
```
